2. Word Embedding Mini Task


```python
import gensim.downloader as api
w2v = api.load("word2vec-google-news-300")
```

```python
w2v.most_similar("king")
```

```python
w2v.most_similar(positive=['king','woman'], negative=['man'])[0]
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

words = ["king","queen","man","woman","apple","orange","car","bus","cat","dog"]

vectors = [w2v[word] for word in words]

pca = PCA(n_components=2)
result = pca.fit_transform(vectors)

plt.figure()
for i, word in enumerate(words):
    plt.scatter(result[i,0], result[i,1])
    plt.text(result[i,0]+0.01, result[i,1]+0.01, word)

plt.show()
```

```python
glove = api.load("glove-wiki-gigaword-100")
glove.most_similar("king")
```

Compare Word2Vec vs GloVe similarity result

```python
w2v_similar = w2v.most_similar("king", topn=10)
glove_similar = glove.most_similar("king", topn=10)
print(w2v_similar)
print(glove_similar)
```
