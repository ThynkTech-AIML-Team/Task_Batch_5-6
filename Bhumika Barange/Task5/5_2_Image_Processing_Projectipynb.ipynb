```python
image = cv2.imread("car image (1).jpg")

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
```

```python
# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Edge Detection
edges = cv2.Canny(gray,100,200)

plt.imshow(edges,cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")
```

```python
# Thresholding
_,threshold=cv2.threshold(gray,127,255,cv2.THRESH_BINARY)

plt.imshow(threshold,cmap='gray')
plt.title("Binary Threshold")
plt.axis("off")
```

```python
flip=cv2.flip(image_rgb,1)

plt.imshow(flip)
plt.title("Flipped Image")
plt.axis("off")
```

```python
(h,w)=image.shape[:2]

center=(w//2,h//2)

matrix=cv2.getRotationMatrix2D(center,45,1)

rotated=cv2.warpAffine(image_rgb,matrix,(w,h))

plt.imshow(rotated)
plt.title("Rotated Image")
plt.axis("off")
```

```python
bright=cv2.convertScaleAbs(image_rgb,alpha=1.2,beta=50)

plt.imshow(bright)
plt.title("Brightness Adjusted")
plt.axis("off")
```
