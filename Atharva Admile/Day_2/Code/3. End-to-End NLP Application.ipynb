**3. End-to-End NLP Application (Choose One):**

• Multi-lingual Sentiment Analysis → Compare sentiment analysis in English vs Hindi
using Hugging Face models.

```python
from transformers import pipeline

# Load models
english_sentiment = pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english')
hindi_sentiment = pipeline('sentiment-analysis', model='LondonStory/txlm-roberta-hindi-sentiment')

# Sample texts
english_texts = [
    "I love this product, it's amazing!",
    "This is the worst experience ever.",
    "Neutral opinion here."
]
hindi_texts = [
    "यह उत्पाद मुझे बहुत पसंद है, यह अद्भुत है!",  # Positive
    "यह सबसे खराब अनुभव है।",  # Negative
    "यहाँ तटस्थ राय है।"  # Neutral
]

# Analyze English
print("English Sentiment:")
for text in english_texts:
    result = english_sentiment(text)[0]
    print(f"Text: {text} -> Label: {result['label']}, Score: {result['score']:.4f}")

# Analyze Hindi
print("\nHindi Sentiment:")
for text in hindi_texts:
    result = hindi_sentiment(text)[0]
    print(f"Text: {text} -> Label: {result['label']}, Score: {result['score']:.4f}")

# Comparison: English models like DistilBERT are highly accurate due to extensive English training data. Hindi models like this one perform well on Devanagari text but may vary in accuracy for dialects, slang, or mixed-language (Hinglish) input. For better results on mixed content, consider multilingual models or fine-tuning.
```
