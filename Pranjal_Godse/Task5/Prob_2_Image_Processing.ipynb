## Step 1: Install and Import Required Libraries

```python

!pip install opencv-python

import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

## Step 2: Upload and Display Image

```python

uploaded = files.upload()

image_path = list(uploaded.keys())[0]

image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```


## Edge Detection (Canny)

Canny Edge Detection is used to detect edges in images by identifying areas with strong intensity gradients.


```python

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

edges = cv2.Canny(gray, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()
```


## Image Thresholding

Thresholding converts a grayscale image into a binary image based on a threshold value.


```python

_, thresh_binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh_binary, cmap='gray')
plt.title("Binary Threshold")
plt.axis("off")
plt.show()
```


## Image Augmentation

Image augmentation helps increase dataset diversity by applying transformations.

### A. Horizontal Flip


```python

flip_horizontal = cv2.flip(image_rgb, 1)

plt.imshow(flip_horizontal)
plt.title("Horizontal Flip")
plt.axis("off")
plt.show()
```

### B. Rotate Image (45 Degrees)

```python

(h, w) = image_rgb.shape[:2]
center = (w // 2, h // 2)

matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(image_rgb, matrix, (w, h))

plt.imshow(rotated)
plt.title("Rotated Image (45 Degrees)")
plt.axis("off")
plt.show()
```

### C. Brightness Adjustment

```python

bright = cv2.convertScaleAbs(image_rgb, alpha=1.2, beta=50)

plt.imshow(bright)
plt.title("Brightness Increased")
plt.axis("off")
plt.show()
```
