```python
import gensim.downloader as api

print("Loading pretrained GloVe model...")

model = api.load("glove-wiki-gigaword-50")

print("Model loaded successfully!")
```

```python
print("Vocabulary size:", len(model.index_to_key))

print("Example words:", model.index_to_key[:10])
```

```python
print("Similar words to 'king':")

similar_words = model.most_similar("king", topn=10)

for word, score in similar_words:
    
    print(word, score)
```

```python
print("king - man + woman =")

result = model.most_similar(
    positive=["king", "woman"],
    negative=["man"]
)

for word, score in result:
    
    print(word, score)
```

```python
print("Paris - France + Italy =")

print(model.most_similar(
    positive=["paris", "italy"],
    negative=["france"]
)[0])
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

words = [
    "king", "queen", "man", "woman",
    "computer", "laptop", "phone",
    "movie", "film", "actor",
    "good", "bad", "happy", "sad"
]

vectors = [model[word] for word in words]

pca = PCA(n_components=2)

result = pca.fit_transform(vectors)

plt.figure(figsize=(10,8))

for i, word in enumerate(words):
    
    plt.scatter(result[i,0], result[i,1])
    
    plt.text(result[i,0]+0.02, result[i,1]+0.02, word)

plt.title("Word Embeddings Visualization using PCA")

plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")

plt.show()
```

```python
print("Similarity between king and queen:",
      model.similarity("king", "queen"))

print("Similarity between king and computer:",
      model.similarity("king", "computer"))

print("Similarity between good and bad:",
      model.similarity("good", "bad"))
```
