```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
questions = [
    "What is your name?",
    "How can I reset my password?",
    "How to contact support?",
    "What are your working hours?",
    "How do I update my profile?",
    "Where is your office located?",
    "How to create an account?",
    "What payment methods are accepted?",
    "How to cancel my subscription?",
    "How to track my order?",
    "Do you offer refunds?",
    "Is my data secure?",
    "How to change my email address?",
    "Can I upgrade my plan?",
    "How long does delivery take?"
]

answers = [
    "I am your FAQ chatbot assistant.",
    "You can reset your password by clicking on 'Forgot Password' at login.",
    "You can contact support via email at support@example.com.",
    "Our working hours are 9 AM to 6 PM, Monday to Friday.",
    "You can update your profile in the account settings section.",
    "Our office is located in Pune, India.",
    "Click on 'Sign Up' and fill in the required details.",
    "We accept credit cards, debit cards, and UPI.",
    "You can cancel your subscription from the billing section.",
    "You can track your order from the orders page.",
    "Yes, we offer refunds within 7 days of purchase.",
    "Yes, we use encryption to secure your data.",
    "You can change your email in profile settings.",
    "Yes, you can upgrade from the subscription page.",
    "Delivery usually takes 3-5 business days."
]
```

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(questions)
```

```python
def chatbot_response(user_query):
    user_vector = vectorizer.transform([user_query])
    similarity_scores = cosine_similarity(user_vector, tfidf_matrix)
    best_match_index = np.argmax(similarity_scores)
    best_score = similarity_scores[0][best_match_index]
    if best_score < 0.3:
        return "Sorry, I did not understand your question."
    return answers[best_match_index]
```

```python
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Chatbot: Goodbye!")
        break
    response = chatbot_response(user_input)
    print("Chatbot:", response)
```
