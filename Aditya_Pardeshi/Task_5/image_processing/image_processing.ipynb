```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("cifar_image_0.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

blur = cv2.GaussianBlur(gray, (5,5), 0)

edges = cv2.Canny(blur, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()
```

```python
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap='gray')
plt.title("Binary Threshold")
plt.axis("off")
plt.show()
```

```python
adaptive = cv2.adaptiveThreshold(
    gray, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    11, 2
)

plt.imshow(adaptive, cmap='gray')
plt.title("Adaptive Threshold")
plt.axis("off")
plt.show()
```

```python
flip_horizontal = cv2.flip(img_rgb, 1)
flip_vertical = cv2.flip(img_rgb, 0)

plt.imshow(flip_horizontal)
plt.title("Horizontal Flip")
plt.axis("off")
plt.show()
```

```python
(h, w) = img.shape[:2]
center = (w // 2, h // 2)

M = cv2.getRotationMatrix2D(center, 45, 1.0)  # 45 degrees
rotated = cv2.warpAffine(img_rgb, M, (w, h))

plt.imshow(rotated)
plt.title("Rotated Image")
plt.axis("off")
plt.show()
```

```python
bright = cv2.convertScaleAbs(img_rgb, alpha=1, beta=50)

plt.imshow(bright)
plt.title("Increased Brightness")
plt.axis("off")
plt.show()
```
