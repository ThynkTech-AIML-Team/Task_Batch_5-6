**Named Entity Recognition Visualizer**

- Use spaCy

- Extract PERSON, ORG, GPE, DATE entities

- Display colored entity output

- Show entity frequency count

```python
import spacy
from spacy import displacy
from collections import Counter

# Load model
nlp = spacy.load('en_core_web_sm')

# Sample text
text = """
Apple Inc. is an American multinational technology company headquartered in Cupertino, California. 
It was founded by Steve Jobs, Steve Wozniak, and Ronald Wayne in 1976. 
On January 3, 2023, Tim Cook announced new products in New York.
The United Nations is headquartered in New York City.
"""

# Process text
doc = nlp(text)

# Extract entities
entities = [(ent.text, ent.label_) for ent in doc.ents if ent.label_ in ['PERSON', 'ORG', 'GPE', 'DATE']]
print("Extracted Entities:", entities)

# Display colored output
displacy.render(doc, style='ent', jupyter=True)

# Entity frequency
freq = Counter([ent[1] for ent in entities])
print("\nEntity Frequency:")
for entity, count in freq.items():
    print(f"{entity}: {count}")
```
