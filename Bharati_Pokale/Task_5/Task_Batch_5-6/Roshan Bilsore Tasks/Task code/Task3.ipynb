```python
#Write a Python function that takes a paragraph and outputs:
# List of sentences
```

```python
import re
```

```python
def analyze_text(text):
    
  
    paragraphs = [p for p in text.split("\n") if p.strip() != ""]
    # Sentences (split using punctuation)
    sentences = re.split(r'[.!?]+', text)
    sentences = [s.strip() for s in sentences if s.strip() != ""]
    # Tokens / words (only words, ignore punctuation)
    tokens = re.findall(r'\b\w+\b', text)
    # Counts
    token_count = len(tokens)
    sentence_count = len(sentences)
    paragraph_count = len(paragraphs)
    return {
        "sentences": sentences,
        "tokens": tokens,
        "token_count": token_count,
        "sentence_count": sentence_count,
        "paragraph_count": paragraph_count
    }
```

```python
text = """Python is a powerful programming language. It is widely used in AI and data science.
It is easy to learn and flexible."""
```

```python
analyze_text(text)
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
def analyze_text_nlp(text):
    sentences = sent_tokenize(text)
    tokens = word_tokenize(text)

    paragraphs = [p for p in text.split("\n") if p.strip() != ""]
    
    return {
        "sentences": sentences,
        "tokens": tokens,
        "token_count": len(tokens),
        "sentence_count": len(sentences),
        "paragraph_count": len(paragraphs)
    }
```

```python
text = """Python is a powerful programming language. It is widely used in AI and data science.
It is easy to learn and flexible! Many developers love it."""
```

```python
result = analyze_text_nlp(text)
result
```
