```python
#import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#loading image
image = cv2.imread("D:\DATA\image_1.jpg")

# Convert BGR to RGB (for correct display in matplotlib)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
#edge detection using Canny
# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply Canny Edge Detection
edges = cv2.Canny(gray, threshold1=100, threshold2=200)

plt.imshow(edges, cmap="gray")
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()
```

```python
#image thresholding
_, thresh_binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh_binary, cmap="gray")
plt.title("Binary Threshold")
plt.axis("off")
plt.show()
```

```python
adaptive_thresh = cv2.adaptiveThreshold(
    gray, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    11, 2
)

plt.imshow(adaptive_thresh, cmap="gray")
plt.title("Adaptive Threshold")
plt.axis("off")
plt.show()
```

```python
#image augmentation
# 1.flip image
flip_horizontal = cv2.flip(image_rgb, 2)

plt.imshow(flip_horizontal)
plt.title("Horizontal Flip")
plt.axis("off")
plt.show()
```

```python
# 2.rotate image
(h, w) = image.shape[:2]
center = (w // 2, h // 2)

matrix = cv2.getRotationMatrix2D(center, 90, 1.0)
rotated = cv2.warpAffine(image_rgb, matrix, (w, h))

plt.imshow(rotated)
plt.title("Rotated Image (90)")
plt.axis("off")
plt.show()
```

```python
# 3.adjusting brightness
# Increase brightness
brightness = 50
bright_image = cv2.convertScaleAbs(image_rgb, alpha=1, beta=brightness)

plt.imshow(bright_image)
plt.title("Brightness Increased")
plt.axis("off")
plt.show()
```
