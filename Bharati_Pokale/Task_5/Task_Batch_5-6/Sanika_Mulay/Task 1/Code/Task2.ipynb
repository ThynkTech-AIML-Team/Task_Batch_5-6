```python
import nltk
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
nltk.download('punkt')
```

```python
def analyze_text(paragraph):
    # Paragraph count
    paragraphs = paragraph.split("\n")
    paragraph_count = len([p for p in paragraphs if p.strip() != ""])
    
    # Sentence tokenization
    sentences = sent_tokenize(paragraph)
    
    # Word tokenization
    tokens = word_tokenize(paragraph)
    
    # Remove punctuation tokens (optional improvement)
    words = [word for word in tokens if word.isalnum()]
    
    print("List of Sentences:")
    print(sentences)
    print("\nList of Word Tokens:")
    print(words)
    print("\nCounts:")
    print("Number of Paragraphs:", paragraph_count)
    print("Number of Sentences:", len(sentences))
    print("Number of Tokens:", len(words))
    
```

```python
text = """Natural Language Processing (NLP) is a branch of Artificial Intelligence.
It helps computers understand human language.

NLP is used in chatbots, translation systems, and sentiment analysis."""
```

```python
analyze_text(text)
```
