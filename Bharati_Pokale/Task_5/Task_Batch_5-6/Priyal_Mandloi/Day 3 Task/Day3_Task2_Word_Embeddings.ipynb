```python
import gensim
import gensim.downloader as api
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
word2vec = api.load("word2vec-google-news-300")
```

```python
word2vec.most_similar("king")
```

```python
word2vec.most_similar(
    positive=['king', 'woman'],
    negative=['man']
)
```

```python
words = [
    'king', 'queen', 'man', 'woman',
    'apple', 'banana', 'orange',
    'car', 'bus', 'train'
]

vectors = [word2vec[word] for word in words]

pca = PCA(n_components=2)
result = pca.fit_transform(vectors)

plt.figure(figsize=(8, 6))
plt.scatter(result[:, 0], result[:, 1])

for i, word in enumerate(words):
    plt.annotate(word, (result[i, 0], result[i, 1]))

plt.title("Word Embeddings Visualization using PCA")
plt.show()
```

## Task 2: Word Embeddings – Conclusion

This task demonstrated the use of pretrained Word2Vec embeddings to capture semantic relationships between words.
The analogy task (king − man + woman) correctly returned "queen", proving meaningful vector representations.
PCA visualization showed semantic clustering of related words.
