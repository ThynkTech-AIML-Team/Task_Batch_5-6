```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

words = ["king","queen","man","woman","prince","princess",
         "boy","girl","father","mother"]

w2v_vectors = [w2v[word] for word in words]
w2v_pca = PCA(n_components=2).fit_transform(w2v_vectors)

glove_vectors = [glove[word] for word in words]
glove_pca = PCA(n_components=2).fit_transform(glove_vectors)

fig, axes = plt.subplots(1, 2, figsize=(14,6))

for i, word in enumerate(words):
    axes[0].scatter(w2v_pca[i,0], w2v_pca[i,1])
    axes[0].text(w2v_pca[i,0], w2v_pca[i,1], word)
axes[0].set_title("Word2Vec Embeddings (PCA)")

for i, word in enumerate(words):
    axes[1].scatter(glove_pca[i,0], glove_pca[i,1])
    axes[1].text(glove_pca[i,0], glove_pca[i,1], word)
axes[1].set_title("GloVe Embeddings (PCA)")

plt.tight_layout()
plt.show()


```

```python
import gensim.downloader as api
w2v = api.load("word2vec-google-news-300")
glove = api.load("glove-wiki-gigaword-100")
```

```python
print("Word2Vec Similar Words:")
print(w2v.most_similar("king"))

print("\nGloVe Similar Words:")
print(glove.most_similar("king"))
```

```python
print("Word2Vec Analogy:")
print(w2v.most_similar(positive=['king','woman'], negative=['man']))

print("\nGloVe Analogy:")
print(glove.most_similar(positive=['king','woman'], negative=['man']))
```

```python
print("Word2Vec similarity (king, queen):",
      w2v.similarity("king", "queen"))

print("GloVe similarity (king, queen):",
      glove.similarity("king", "queen"))
```
