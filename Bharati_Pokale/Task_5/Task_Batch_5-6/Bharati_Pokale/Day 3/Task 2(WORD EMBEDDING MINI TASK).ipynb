```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Load Pretrained Model
```

```python
import gensim.downloader as api

model = api.load("glove-wiki-gigaword-50")

```

```python
# Similar Words
```

```python
print("Similar words to 'king':")
print(model.most_similar("king", topn=5))

```

```python
# Analogy
```

```python
print("king - man + woman:")
print(model.most_similar(positive=["king", "woman"], negative=["man"], topn=5))
```

```python
# PCA Visualization
```

```python

words = ["king", "queen", "man", "woman",
         "paris", "france", "london", "england",
         "apple", "orange"]

vectors = np.array([model[word] for word in words])

pca = PCA(n_components=2)
reduced = pca.fit_transform(vectors)

plt.figure()
for i, word in enumerate(words):
    plt.scatter(reduced[i, 0], reduced[i, 1])
    plt.text(reduced[i, 0], reduced[i, 1], word)

plt.title("Word Embedding PCA Visualization")
plt.show()
```
