```python
import nltk
import string

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize

nltk.download('punkt')
nltk.download('stopwords')
```

```python
text = """
Artificial Intelligence is transforming the world rapidly. It is being used in healthcare,
finance, transportation, education, and many other industries. AI helps automate repetitive
tasks, improve decision making, and increase efficiency. Machine learning, a subset of AI,
allows systems to learn from data and improve over time. Deep learning enables advanced
applications such as image recognition and natural language processing. As AI continues to
develop, it is expected to create new opportunities and challenges for society.
"""

print("Original Text:\n")
print(text)
```

```python
stop_words = set(stopwords.words("english"))

words = word_tokenize(text.lower())

words = [
    word for word in words
    if word not in stop_words
    and word not in string.punctuation
]
```

```python
word_freq = {}

for word in words:
    
    if word not in word_freq:
        
        word_freq[word] = 1
        
    else:
        
        word_freq[word] += 1

word_freq
```

```python
sentences = sent_tokenize(text)

sentence_scores = {}

for sentence in sentences:
    
    for word in word_tokenize(sentence.lower()):
        
        if word in word_freq:
            
            if sentence not in sentence_scores:
                
                sentence_scores[sentence] = word_freq[word]
                
            else:
                
                sentence_scores[sentence] += word_freq[word]

sentence_scores
```

```python
import heapq

summary_sentences = heapq.nlargest(
    2,
    sentence_scores,
    key=sentence_scores.get
)

summary = " ".join(summary_sentences)

print("Summary:\n")

print(summary)
```

```python
print("Original length:", len(text.split()))

print("Summary length:", len(summary.split()))
```
