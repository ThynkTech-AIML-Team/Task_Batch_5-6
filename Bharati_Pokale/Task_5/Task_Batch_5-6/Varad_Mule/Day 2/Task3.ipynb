```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import re
from nltk.corpus import stopwords
from sklearn.datasets import fetch_20newsgroups
```

```python
dataset_bundle = fetch_20newsgroups(
    subset='train',
    remove=('headers', 'footers', 'quotes')
)
```

```python
raw_articles = dataset_bundle.data[:2000]
```

```python
stop_words = set(stopwords.words('english'))
```

```python
def preprocess_text(text):
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    text = text.lower()
    tokens = text.split()
    tokens = [word for word in tokens if word not in stop_words and len(word) > 3]
    return " ".join(tokens)
```

```python
processed_articles = [preprocess_text(doc) for doc in raw_articles]
```

```python
vector_space_builder = TfidfVectorizer()
document_matrix = vector_space_builder.fit_transform(processed_articles)
```

```python
def find_similar_articles(query, top_n=3):
    query_cleaned = preprocess_text(query)
    query_vector = vector_space_builder.transform([query_cleaned])
    
    similarity_scores = cosine_similarity(query_vector, document_matrix)
    top_indices = similarity_scores[0].argsort()[-top_n:][::-1]
    
    return [raw_articles[i] for i in top_indices]
```

```python
results = find_similar_articles("machine learning and artificial intelligence")
```

```python
for i, article in enumerate(results):
    print(f"\nResult {i+1}:\n")
    print(article[:500])
```
