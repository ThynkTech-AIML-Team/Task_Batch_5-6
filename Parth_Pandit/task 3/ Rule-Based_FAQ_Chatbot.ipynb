```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
questions = [
    "What is artificial intelligence?",
    "What is machine learning?",
    "What is deep learning?",
    "Difference between AI and ML",
    "What is NLP?",
    "What is supervised learning?",
    "What is unsupervised learning?",
    "What is overfitting?",
    "What is underfitting?",
    "What is a neural network?",
    "What is Python used for?",
    "What is data science?",
    "What is logistic regression?",
    "What is Naive Bayes algorithm?",
    "What is TF-IDF?",
    "What is cosine similarity?"
]

answers = [
    "Artificial Intelligence is the simulation of human intelligence in machines.",
    "Machine Learning is a subset of AI that enables systems to learn from data.",
    "Deep Learning is a subset of ML that uses neural networks with many layers.",
    "AI is the broader concept, while ML is a subset of AI.",
    "NLP is a field of AI that helps computers understand human language.",
    "Supervised learning uses labeled data to train models.",
    "Unsupervised learning works with unlabeled data.",
    "Overfitting occurs when a model learns noise instead of patterns.",
    "Underfitting occurs when a model fails to capture patterns.",
    "A neural network is a model inspired by the human brain.",
    "Python is used for web development, data science, AI, and automation.",
    "Data science involves extracting insights from data.",
    "Logistic regression is a classification algorithm.",
    "Naive Bayes is a probabilistic classification algorithm.",
    "TF-IDF measures the importance of words in a document.",
    "Cosine similarity measures similarity between two vectors."
]
```

```python
vectorizing text:
```

```python
vectorizer = TfidfVectorizer()
question_vectors = vectorizer.fit_transform(questions)
```

```python
Chatbot logic:
```

```python
def chatbot_response(user_query):
    user_vector = vectorizer.transform([user_query])
    similarity_scores = cosine_similarity(user_vector, question_vectors)
    
    best_match_index = np.argmax(similarity_scores)
    best_score = similarity_scores[0][best_match_index]
    
    if best_score < 0.2:
        return "Sorry, I don't understand the question."
    
    return answers[best_match_index]
```

```python
Testing :
```

```python
queries = [
    "Explain artificial intelligence",
    "What is overfitting in machine learning?",
    "Tell me about TF-IDF",
    "What does cosine similarity mean?",
    "What is Python?"
]

for q in queries:
    print("User:", q)
    print("Bot :", chatbot_response(q))
    print("-" * 50)
```

```python
Manual testing questions:
```

```python
while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Chatbot: Goodbye!")
        break
    print("Chatbot:", chatbot_response(user_input))
```
