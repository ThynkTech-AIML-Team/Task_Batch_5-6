```python
import gensim.downloader as api
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np
```

```python
glove = api.load("glove-wiki-gigaword-100")
```

```python
glove.similarity("king", "queen")
```

```python
glove.most_similar("movie")
```

```python
glove.most_similar(
    positive=["king", "woman"],
    negative=["man"]
)
```

```python
words = [
    "king", "queen", "man", "woman",
    "movie", "film", "actor", "actress"
]
```

```python
vectors = np.array([glove[word] for word in words])
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
reduced = pca.fit_transform(vectors)
```

```python
plt.figure(figsize=(8,6))
for i, word in enumerate(words):
    plt.scatter(reduced[i,0], reduced[i,1])
    plt.text(reduced[i,0]+0.01, reduced[i,1]+0.01, word)

plt.title("PCA Visualization of GloVe Word Embeddings")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
import gensim.downloader as api

glove = api.load("glove-wiki-gigaword-100")
word2vec = api.load("word2vec-google-news-300")
fasttext = api.load("fasttext-wiki-news-subwords-300")
```

```python
oov_word = "runningly"

print("Word2Vec OOV check:", oov_word in word2vec.key_to_index)
print("GloVe OOV check:", oov_word in glove.key_to_index)
print("FastText OOV check:", oov_word in fasttext.key_to_index)
```

### Out-of-Vocabulary (OOV) Handling Comparison

The word *"runningly"* is not present in the vocabularies of Word2Vec,
GloVe, or the FastText KeyedVectors used in this experiment.

Word2Vec and GloVe are unable to represent unseen words because they rely
on fixed word-level embeddings.

FastText, however, is designed to generate embeddings using character
n-grams, allowing it to conceptually handle OOV words. Due to limitations
of Gensims pretrained KeyedVectors interface, subword-based vectors are
not directly accessible, but FastText still provides superior robustness
to unseen words in theory and practice.
