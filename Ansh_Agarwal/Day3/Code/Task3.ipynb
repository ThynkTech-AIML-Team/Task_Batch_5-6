```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
faq_data = {
    "What is your name?": "I am a simple FAQ chatbot.",
    "How are you?": "I am functioning properly!",
    "What is AI?": "AI stands for Artificial Intelligence.",
    "What is Machine Learning?": "Machine Learning is a subset of AI.",
    "What is Deep Learning?": "Deep Learning is a subset of Machine Learning.",
    "What is Python?": "Python is a programming language.",
    "What is NLP?": "NLP stands for Natural Language Processing.",
    "What is data science?": "Data science involves extracting insights from data.",
    "What is TF-IDF?": "TF-IDF measures word importance in text.",
    "What is cosine similarity?": "Cosine similarity measures similarity between vectors.",
    "What is overfitting?": "Overfitting occurs when a model memorizes training data.",
    "What is underfitting?": "Underfitting occurs when a model fails to learn patterns.",
    "What is supervised learning?": "Supervised learning uses labeled data.",
    "What is unsupervised learning?": "Unsupervised learning finds patterns in unlabeled data.",
    "What is classification?": "Classification is predicting categories.",
    "What is regression?": "Regression predicts continuous values."
}
```

```python
questions = list(faq_data.keys())

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(questions)
```

```python
vectorizer = TfidfVectorizer(
    lowercase=True,
    stop_words='english'
)

tfidf_matrix = vectorizer.fit_transform(questions)

def chatbot_response(user_query):
    user_tfidf = vectorizer.transform([user_query])
    similarity = cosine_similarity(user_tfidf, tfidf_matrix)
    
    index = np.argmax(similarity)
    score = similarity[0][index]
    
    if score < 0.15:
        return "Sorry, I do not understand your question."
    
    return faq_data[questions[index]]
```

```python
while True:
    user_input = input("You: ")
    
    if user_input.lower() == "exit":
        print("Chatbot: Goodbye!")
        break
    
    response = chatbot_response(user_input)
    print("Chatbot:", response)
```
