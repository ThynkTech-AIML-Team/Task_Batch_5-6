```python
import nltk
nltk.download('punkt')
```

```python
def analyze_text(paragraph):
    sentences = nltk.sent_tokenize(paragraph)
    tokens = nltk.word_tokenize(paragraph)

    print("Sentences:", sentences)
    print("Tokens:", tokens)
    print("Number of sentences:", len(sentences))
    print("Number of tokens:", len(tokens))
    print("Number of paragraphs:", paragraph.count("\n") + 1)
```

```python
text = """Artificial Intelligence (AI) is transforming the modern world at a rapid pace. It refers to the simulation of human intelligence in machines that are programmed to think and learn. 
AI systems can analyze large amounts of data, recognize patterns, and make decisions with minimal human intervention. 
From healthcare to finance, AI is being applied across multiple industries to improve efficiency and accuracy. 
In healthcare, AI helps doctors diagnose diseases by analyzing medical images and patient records. 
In finance, AI is used for fraud detection, risk assessment, and automated trading systems."""
```

```python
analyze_text(text)
```
