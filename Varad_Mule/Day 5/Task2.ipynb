```python
(h, w) = image.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((w/2, h/2), 45, 1)
rotated = cv2.warpAffine(image_rgb, rotation_matrix, (w, h))
```

```python
plt.imshow(rotated)
plt.title("Rotated Image (45 degrees)")
plt.axis("off")
plt.show()
```

```python
bright = cv2.convertScaleAbs(image_rgb, alpha=1, beta=50)
```
