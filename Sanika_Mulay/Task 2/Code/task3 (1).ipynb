##### End-to-End NLP Application
 
##### News Similarity Search â†’ Input one article, output top 3 most similar ones

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
print("Loading NLP Model...")
model = SentenceTransformer('all-MiniLM-L6-v2')
print("Model Loaded!\n")
```

```python
news_articles = [
    "India wins cricket world cup after thrilling final match.",
    "Stock market crashes due to global economic slowdown.",
    "New AI technology is transforming healthcare industry.",
    "Heavy rainfall causes floods in several cities.",
    "SpaceX launches new satellite into orbit successfully.",
    "Government announces new education policy reforms."
]
```

```python
news_embeddings = model.encode(news_articles)
```

```python
input_article = input("Enter a News Article: ")
```

```python
input_embedding = model.encode([input_article])
```

```python
similarities = cosine_similarity(input_embedding, news_embeddings)[0]
```

```python
top_indices = np.argsort(similarities)[-3:][::-1]
```

```python

print("\nTop 3 Similar News Articles:\n")
for idx in top_indices:
    print(f"Article: {news_articles[idx]}")
    print(f"Similarity Score: {similarities[idx]:.4f}\n")
```
