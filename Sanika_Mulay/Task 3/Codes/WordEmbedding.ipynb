```python
import gensim.downloader as api
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import numpy as np
```

```python
# Load Word2Vec 
model_w2v = api.load("word2vec-google-news-300")

# Load GloVe 
model_glove = api.load("glove-wiki-gigaword-100")

print("Both Models Loaded Successfully!")
```

```python
print("Word2Vec - Similar to 'king':")
print(model_w2v.most_similar("king"))

print("\nGloVe - Similar to 'king':")
print(model_glove.most_similar("king"))

```

```python
print("Word2Vec Analogy:")
print(model_w2v.most_similar(positive=['king','woman'], negative=['man']))

print("\nGloVe Analogy:")
print(model_glove.most_similar(positive=['king','woman'], negative=['man']))

```

```python
print("Word2Vec Similarity (king, queen):",
      model_w2v.similarity("king", "queen"))

print("GloVe Similarity (king, queen):",
      model_glove.similarity("king", "queen"))
```

```python
words = ["king", "queen", "man", "woman",
         "prince", "princess", "boy", "girl",
         "father", "mother", "throne",
         "monarch", "child", "royal", "crown"]

# Choose which model to visualize
model = model_w2v   # change to model_glove if needed

word_vectors = np.array([model[word] for word in words])

pca = PCA(n_components=2)
result = pca.fit_transform(word_vectors)

plt.figure()
plt.scatter(result[:, 0], result[:, 1])

for i, word in enumerate(words):
    plt.annotate(word, xy=(result[i, 0], result[i, 1]))

plt.title("Word Embedding Visualization using PCA")
plt.show()
```
