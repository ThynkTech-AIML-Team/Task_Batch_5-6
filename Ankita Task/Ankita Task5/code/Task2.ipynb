```python
import cv2
import matplotlib.pyplot as plt
import os

# Load Image
image_path = "image.jpg"

if not os.path.exists(image_path):
    print("Image not found!")
    exit()

image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image_rgb, cv2.COLOR_RGB2GRAY)

# Apply Canny Edge Detection
edges = cv2.Canny(gray, 100, 200)

# Display
plt.figure(figsize=(8,4))

plt.subplot(1,2,1)
plt.imshow(image_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(edges, cmap="gray")
plt.title("Canny Edges")
plt.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import os

# Load Image
image_path = "image.jpg"

if not os.path.exists(image_path):
    print("Image not found!")
    exit()

image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Thresholding
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
adaptive = cv2.adaptiveThreshold(
    gray, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    11, 2
)
_, otsu = cv2.threshold(
    gray, 0, 255,
    cv2.THRESH_BINARY + cv2.THRESH_OTSU
)

# Display
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.imshow(gray, cmap="gray")
plt.title("Grayscale")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(binary, cmap="gray")
plt.title("Binary")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(adaptive, cmap="gray")
plt.title("Adaptive")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(otsu, cmap="gray")
plt.title("Otsu")
plt.axis("off")

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import os

# Load Image
image_path = "image.jpg"

if not os.path.exists(image_path):
    print("Image not found!")
    exit()

image = cv2.imread(image_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Flip
flip_horizontal = cv2.flip(image_rgb, 1)

# Rotate 45 degrees
h, w = image_rgb.shape[:2]
matrix = cv2.getRotationMatrix2D((w/2, h/2), 45, 1)
rotated = cv2.warpAffine(image_rgb, matrix, (w, h))

# Brightness Adjustment
bright = cv2.convertScaleAbs(image_rgb, alpha=1, beta=50)

# Display
plt.figure(figsize=(10,6))

plt.subplot(2,2,1)
plt.imshow(image_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(2,2,2)
plt.imshow(flip_horizontal)
plt.title("Flip Horizontal")
plt.axis("off")

plt.subplot(2,2,3)
plt.imshow(rotated)
plt.title("Rotated 45")
plt.axis("off")

plt.subplot(2,2,4)
plt.imshow(bright)
plt.title("Brightness Increased")
plt.axis("off")

plt.show()
```
