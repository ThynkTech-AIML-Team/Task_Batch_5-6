Word Embedding Mini Task

```python
import gensim.downloader as api

model = api.load("word2vec-google-news-300")
```

```python
print("similar words to 'king':")
print(model.most_similar("king"))
```

```python
print("basic analogy (king - man + woman)")
print(model.most_similar(positive=['king','woman'], negative=['man']))
```

```python
print(" Visualize 10–20 words using PCA (2D plot)")
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

words = ['king','queen','man','woman','apple','orange','dog','cat','car','bus']

vectors = [model[word] for word in words]

pca = PCA(n_components=2)
result = pca.fit_transform(vectors)

for i, word in enumerate(words):
    plt.scatter(result[i,0], result[i,1])
    plt.text(result[i,0]+0.01, result[i,1]+0.01, word)

plt.show()
```
