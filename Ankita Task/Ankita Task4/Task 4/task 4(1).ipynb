1. Exploratory Data Analysis (EDA) Project: 
This EDA project cleans, manipulates, and visualizes a titanic dataset to uncover patterns, handling missing values via imputation/dropping, engineering new features, and using Seaborn/Matplotlib for insights. Key steps include identifying distributions, correlations, and anomalies to inform predictive modeling.

```python
# Required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("datasets/Titanic-Dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#missing values
df.isnull().sum()
```

```python
# Fill Age with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill Embarked with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Drop unnecessary columns
df = df.drop(columns=['PassengerId', 'Name', 'Ticket'])
```

```python
df.columns
```

```python
#convert into numeric
df['Sex'] = df['Sex'].map({'male':0, 'female':1})
```

```python
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)
```

```python
#data visualization
df.hist(figsize=(10,8))
plt.show()
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```
