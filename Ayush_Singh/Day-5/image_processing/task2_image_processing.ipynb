```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Path
image_path = "input_images/sample.jpg"

# Read image
image = cv2.imread(image_path)

# Convert BGR to RGB for display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Canny Edge Detection
edges = cv2.Canny(gray, threshold1=100, threshold2=200)

plt.imshow(edges, cmap="gray")
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()

# Save
cv2.imwrite("output_images/canny_edges.jpg", edges)
```

```python
# Simple Binary Threshold
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap="gray")
plt.title("Binary Threshold")
plt.axis("off")
plt.show()

cv2.imwrite("output_images/threshold.jpg", thresh)
```

```python
flip_horizontal = cv2.flip(image_rgb, 1)

plt.imshow(flip_horizontal)
plt.title("Flipped Image")
plt.axis("off")
plt.show()

cv2.imwrite("output_images/flipped.jpg", cv2.cvtColor(flip_horizontal, cv2.COLOR_RGB2BGR))
```

```python
(h, w) = image.shape[:2]
center = (w // 2, h // 2)

matrix = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(image_rgb, matrix, (w, h))

plt.imshow(rotated)
plt.title("Rotated Image (45 degrees)")
plt.axis("off")
plt.show()

cv2.imwrite("output_images/rotated.jpg", cv2.cvtColor(rotated, cv2.COLOR_RGB2BGR))
```

```python
bright = cv2.convertScaleAbs(image_rgb, alpha=1.2, beta=40)

plt.imshow(bright)
plt.title("Brightness Adjusted")
plt.axis("off")
plt.show()

cv2.imwrite("output_images/brightness.jpg", cv2.cvtColor(bright, cv2.COLOR_RGB2BGR))
```
