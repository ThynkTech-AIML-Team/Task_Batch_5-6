```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 2: Read Image
```

```python
image = cv2.imread("sample.jpg")

# Convert BGR to RGB (for correct display in matplotlib)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
# Edge Detection  Canny
```

```python
#  Convert to Grayscale
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Apply Canny
```

```python
edges = cv2.Canny(gray, threshold1=100, threshold2=200)

plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()
```

```python
# Image Thresholding
```

```python
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap='gray')
plt.title("Binary Threshold")
plt.axis("off")
plt.show()
```

```python
# Adaptive Threshold (Better)
adaptive = cv2.adaptiveThreshold(
    gray, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    11, 2
)

plt.imshow(adaptive, cmap='gray')
plt.title("Adaptive Threshold")
plt.axis("off")
plt.show()
```

```python
# Image Augmentation
```

```python
# A) Flip Image
```

```python

flip_horizontal = cv2.flip(image_rgb, 1)
flip_vertical = cv2.flip(image_rgb, 0)

plt.imshow(flip_horizontal)
plt.title("Horizontal Flip")
plt.axis("off")
plt.show()
```

```python
# B) Rotate Image
```

```python
(h, w) = image.shape[:2]
center = (w//2, h//2)

M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(image_rgb, M, (w, h))

plt.imshow(rotated)
plt.title("Rotated 45 Degrees")
plt.axis("off")
plt.show()
```

```python
# C) Brightness Adjustment
```

```python
bright = cv2.convertScaleAbs(image_rgb, alpha=1, beta=50)

plt.imshow(bright)
plt.title("Brightness Increased")
plt.axis("off")
plt.show()
```
