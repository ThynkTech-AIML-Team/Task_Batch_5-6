```python
# Section 1: Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os

```

```python

# Section 2: Verify Working Directory
print("Current Folder:")
print(os.getcwd())

print("\nFiles in this folder:")
print(os.listdir())

```

```python

# Section 3: Load Image
image_path = "sample.jpg"

img = cv2.imread(image_path)

if img is None:
    print("\nERROR: Image not found.")
    print("Make sure sample.jpg is inside this folder:")
    print(os.getcwd())

else:
    print("\nImage loaded successfully!")


    # Convert BGR to RGB
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    plt.imshow(img_rgb)
    plt.title("Original Image")
    plt.axis("off")
    plt.show()


    # Section 4: Convert to Grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    plt.imshow(gray, cmap="gray")
    plt.title("Grayscale Image")
    plt.axis("off")
    plt.show()


    # Section 5: Edge Detection (Canny)
    edges = cv2.Canny(gray, 100, 200)

    plt.imshow(edges, cmap="gray")
    plt.title("Edge Detection")
    plt.axis("off")
    plt.show()


    # Section 6: Thresholding
    ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

    plt.imshow(thresh, cmap="gray")
    plt.title("Binary Threshold")
    plt.axis("off")
    plt.show()


    # Section 7: Flip Image
    flip = cv2.flip(img_rgb, 1)

    plt.imshow(flip)
    plt.title("Flipped Image")
    plt.axis("off")
    plt.show()


    # Section 8: Rotate Image
    (h, w) = img.shape[:2]
    center = (w // 2, h // 2)

    matrix = cv2.getRotationMatrix2D(center, 45, 1.0)

    rotate = cv2.warpAffine(img_rgb, matrix, (w, h))

    plt.imshow(rotate)
    plt.title("Rotated Image (45)")
    plt.axis("off")
    plt.show()


    # Section 9: Brightness Adjustment
    bright = cv2.convertScaleAbs(img_rgb, alpha=1, beta=50)

    plt.imshow(bright)
    plt.title("Brightness Increased")
    plt.axis("off")
    plt.show()

    print("\nTask 2 completed successfully!")
```
