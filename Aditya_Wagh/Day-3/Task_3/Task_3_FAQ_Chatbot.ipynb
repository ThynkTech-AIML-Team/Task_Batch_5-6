```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
questions = [

    "What is Python?",
    "What is machine learning?",
    "What is artificial intelligence?",
    "What is deep learning?",
    "What is NLP?",
    "What is data science?",
    "What is chatbot?",
    "What is TF-IDF?",
    "What is cosine similarity?",
    "What is Word2Vec?",
    "What is GloVe?",
    "What is FastText?",
    "What is supervised learning?",
    "What is unsupervised learning?",
    "What is classification?",
    "What is regression?",
    "What is clustering?",
    "What is neural network?",
    "What is spaCy?",
    "What is scikit-learn?"

]

answers = [

    "Python is a programming language used in AI and data science.",
    "Machine learning allows computers to learn from data.",
    "Artificial intelligence simulates human intelligence.",
    "Deep learning uses neural networks for learning.",
    "NLP helps computers understand human language.",
    "Data science extracts insights from data.",
    "Chatbot is a program that talks with users.",
    "TF-IDF converts text into numerical features.",
    "Cosine similarity measures similarity between texts.",
    "Word2Vec converts words into vectors.",
    "GloVe is a pretrained word embedding model.",
    "FastText handles unknown words using subwords.",
    "Supervised learning uses labeled data.",
    "Unsupervised learning uses unlabeled data.",
    "Classification predicts categories.",
    "Regression predicts numbers.",
    "Clustering groups similar data.",
    "Neural network is a brain-inspired model.",
    "spaCy is an NLP library.",
    "scikit-learn is a machine learning library."

]

print("Total FAQ Questions:", len(questions))
```

```python
vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(questions)

print("TF-IDF Matrix created successfully")
```

```python
history = []

print("\nFAQ Chatbot Ready!")
print("Type 'exit' to stop.\n")

while True:

    user_input = input("You: ")

    if user_input.lower() == "exit":
        print("\nFull Chat History:")
        for q, a in history:
            print(f"Q: {q}")
            print(f"A: {a}\n")
        break

    user_vector = vectorizer.transform([user_input])

    similarity = cosine_similarity(user_vector, tfidf_matrix)

    best_match_index = similarity.argmax()

    best_score = similarity[0][best_match_index]

    threshold = 0.3

    if best_score > threshold:
        bot_answer = answers[best_match_index]
    else:
        bot_answer = "Sorry, I don't understand that question."

    print("Bot:", bot_answer)

    history.append((user_input, bot_answer))

    print("\nChat History:")
    for q, a in history:
        print(f"Q: {q}")
        print(f"A: {a}")
    print()
```
