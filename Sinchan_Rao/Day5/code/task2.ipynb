```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("sample.jpg")

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
edges = cv2.Canny(gray, 100, 200)

plt.imshow(edges, cmap='gray')
plt.title("Canny Edge Detection")
plt.axis("off")
plt.show()
```

```python
ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

plt.imshow(thresh, cmap='gray')
plt.title("Binary Thresholding")
plt.axis("off")
plt.show()
```

```python
flip_horizontal = cv2.flip(img, 1)
flip_horizontal = cv2.cvtColor(flip_horizontal, cv2.COLOR_BGR2RGB)

plt.imshow(flip_horizontal)
plt.title("Horizontal Flip")
plt.axis("off")
plt.show()

flip_vertical = cv2.flip(img, 0)
flip_vertical = cv2.cvtColor(flip_vertical, cv2.COLOR_BGR2RGB)

plt.imshow(flip_vertical)
plt.title("Vertical Flip")
plt.axis("off")
plt.show()
```

```python
(h, w) = img.shape[:2]
center = (w // 2, h // 2)

M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(img, M, (w, h))
rotated = cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB)

plt.imshow(rotated)
plt.title("Rotated 45 Degrees")
plt.axis("off")
plt.show()
```

```python
bright = cv2.convertScaleAbs(img, alpha=1, beta=50)
bright = cv2.cvtColor(bright, cv2.COLOR_BGR2RGB)

plt.imshow(bright)
plt.title("Increased Brightness")
plt.axis("off")
plt.show()

dark = cv2.convertScaleAbs(img, alpha=1, beta=-50)
dark = cv2.cvtColor(dark, cv2.COLOR_BGR2RGB)

plt.imshow(dark)
plt.title("Decreased Brightness")
plt.axis("off")
plt.show()
```
