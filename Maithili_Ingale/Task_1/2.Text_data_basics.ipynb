```python
import re

def analyze_text(paragraph):
    # Paragraph count (separated by blank lines)
    paragraphs = [p for p in paragraph.strip().split("\n\n") if p.strip()]
    para_count = len(paragraphs)

    # Sentence split
    sentences = re.split(r'[.!?]+', paragraph)
    sentences = [s.strip() for s in sentences if s.strip()]
    sentence_count = len(sentences)

    # Word tokenization
    tokens = re.findall(r'\b\w+\b', paragraph.lower())
    token_count = len(tokens)

    return {
        "sentences": sentences,
        "tokens": tokens,
        "sentence_count": sentence_count,
        "token_count": token_count,
        "paragraph_count": para_count
    }
```

```python
text = """Hello world! This is a test paragraph. NLP is interesting.

This is the second paragraph for testing."""
```

```python
result = analyze_text(text)

print("Sentences:")
print(result["sentences"])

print("\nTokens:")
print(result["tokens"])

print("\nSentence Count:", result["sentence_count"])
print("Token Count:", result["token_count"])
print("Paragraph Count:", result["paragraph_count"])
```

```python
for key, value in result.items():
    print(f"{key} : {value}")
```
