```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
sentences = [
    "NLP is fun and NLP is useful",
    "I love learning NLP techniques",
    "TF IDF gives importance to rare words"
]
```

```python
bow_vectorizer = CountVectorizer()  # default tokenization + lowercasing
bow_matrix = bow_vectorizer.fit_transform(sentences)

bow_df = pd.DataFrame(
    bow_matrix.toarray(),
    columns=bow_vectorizer.get_feature_names_out()
)

bow_df
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(sentences)

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=tfidf_vectorizer.get_feature_names_out()
)

tfidf_df
```

```python
bow_global = bow_df.sum(axis=0).sort_values(ascending=False)
bow_global.head(15)
```

```python
tfidf_global = tfidf_df.sum(axis=0).sort_values(ascending=False)
tfidf_global.head(15)
```

```python
compare = pd.DataFrame({
    "BoW_total_count": bow_global,
    "TFIDF_total_score": tfidf_global
}).fillna(0)

compare_sorted = compare.sort_values("TFIDF_total_score", ascending=False)
compare_sorted.head(20)
```
