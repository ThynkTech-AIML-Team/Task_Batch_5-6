```python
import gensim.downloader as api
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
w2v=api.load("word2vec-google-news-300")
glove=api.load("glove-wiki-gigaword-100")
```

```python
print(w2v.most_similar("king",topn=5))
print(glove.most_similar("king",topn=5))
```

```python
#bsic analog
print(w2v.most_similar(positive=["king","woman"],negative=["man"],topn=5))
print(glove.most_similar(positive=["king","woman"],negative=["man"],topn=5))
```

```python
words=["king","queen","man","woman","boy","girl","prince","princess","apple","orange","fruit","car","bus","train","city"]
vecs=np.array([w2v[w] for w in words])
pca=PCA(n_components=2)
r=pca.fit_transform(vecs)
plt.figure(figsize=(7,7))
plt.scatter(r[:,0],r[:,1])
for i,w in enumerate(words):
    plt.text(r[i,0],r[i,1],w)
plt.show()
```

```python
for w in ["king","apple","city"]:
    print(w,"W2V",w2v.most_similar(w,topn=3),"GloVe",glove.most_similar(w,topn=3))
```
