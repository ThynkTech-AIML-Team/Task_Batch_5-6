```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/lutzhamel/fake-news/master/data/fake_or_real_news.csv")
df["label"]=df["label"].map({"REAL":1,"FAKE":0})
X_train,X_test,y_train,y_test=train_test_split(df["text"],df["label"],test_size=0.2,random_state=42)
tf=TfidfVectorizer(stop_words="english",max_features=5000)
Xtr=tf.fit_transform(X_train)
Xte=tf.transform(X_test)
```

```python
lr=LogisticRegression(max_iter=1000)
nb=MultinomialNB()
lr.fit(Xtr,y_train)
nb.fit(Xtr,y_train)
for m in [lr,nb]:
    p=m.predict(Xte)
    print(type(m).__name__,accuracy_score(y_test,p))
    print(confusion_matrix(y_test,p))
```
