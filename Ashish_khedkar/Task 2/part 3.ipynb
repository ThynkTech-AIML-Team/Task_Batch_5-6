```python
from transformers import pipeline
from langdetect import detect

# English-only sentiment model
english_sentiment = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

# Multilingual sentiment model (English + Hindi)
multilingual_sentiment = pipeline(
    "sentiment-analysis",
    model="nlptown/bert-base-multilingual-uncased-sentiment"
)
```

```python
def detect_language(text):
    try:
        return detect(text)
    except:
        return "unknown"
```

```python
def analyze_sentiment(text):
    lang = detect_language(text)

    if lang == "en":
        result = english_sentiment(text)[0]
        return {
            "language": "English",
            "model": "English BERT",
            "sentiment": result["label"],
            "confidence": round(result["score"], 3)
        }

    else:
        result = multilingual_sentiment(text)[0]
        stars = int(result["label"][0])

        sentiment = (
            "Negative" if stars <= 2 else
            "Neutral" if stars == 3 else
            "Positive"
        )

        return {
            "language": "Hindi / Other",
            "model": "Multilingual BERT",
            "sentiment": sentiment,
            "confidence": round(result["score"], 3)
        }
```

```python
text_en = "The movie was absolutely amazing and inspiring"
output_en = analyze_sentiment(text_en)
print(output_en)
```

```python
text_hi = "यह फिल्म बहुत खराब थी और मुझे बिल्कुल पसंद नहीं आई"
output_hi = analyze_sentiment(text_hi)
print(output_hi)
```

```python
english_text = "This phone has excellent battery life"
hindi_text = "इस फोन की बैटरी लाइफ बहुत अच्छी है"

print(analyze_sentiment(english_text))
print(analyze_sentiment(hindi_text))
```
